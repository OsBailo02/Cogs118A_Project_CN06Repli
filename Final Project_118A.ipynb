```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('Call_Detail_Report.csv')
df2 = pd.read_csv('All_Data.csv')
```

```python
df1 = df1.drop(columns=['Note Taken', 'Appt Set', 'Lead'])
df1
```

```python
df1['Call Result'] = df1['Call Result'].replace('Contact', 1)
df1['Call Result'] = df1['Call Result'].replace('No Contact', -1)
df1['Call Result'] = df1['Call Result'].replace('Skipped', -1)
df1['Call Result'] = df1['Call Result'].replace('No Answer', -1)
df1['Call Result'] = df1['Call Result'].replace('Disconnected', -1)
df1['Call Result'] = df1['Call Result'].replace('Answering Machine', -1)
df1['Call Result'] = df1['Call Result'].replace('Busy', -1)
df1['Call Result'] = df1['Call Result'].replace('Drop Message', -1)
df1['Call Result'] = df1['Call Result'].replace('Other', -1)
df1['Call Result'] = df1['Call Result'].replace('Bad Number', -1)
df1['Call Result'] = df1['Call Result'].replace('Fax', -1)
df1['Call Result'] = df1['Call Result'].replace('Contact', -1)
df1['Call Result'] = df1['Call Result'].replace('DNC Number', -1)
df1['Call Result'] = df1['Call Result'].replace('Line Error', -1)
df1['Call Result'] = df1['Call Result'].replace('Line Problem', -1)

df1
```
