From a quick glance, we see that the dates are in some variation of YYYY-\*. Fortunately, this makes  things a lot easier for us to standardize. We will use the function below to standardize the date.

```python
def simplify_date(myYear):
    
    myYear = str(myYear).strip()
    
    if '-' in myYear:
        elem = myYear.split('-')
        output = elem[0]
        output.strip()
        
    else:
        output = myYear
    
    if output.isdigit():
        return int(output)
    else:
        return None
```

```python
df_horror['publishedDate'] = df_horror['publishedDate'].apply(simplify_date)
df_romance['publishedDate'] = df_romance['publishedDate'].apply(simplify_date)
```

### 1. Tokenize the descriptions

Using the nltk module and its tokenization functionality, we will process the descrptions for the books in the horror and romance dataframes and break the text down into individual words.

We will now remove stop words so we can focus on words of relevant importance.

```python
stop_words = set(stopwords.words('english'))

df_horror["stopwords_removed"] = df_horror["tokenized_description"].apply(lambda x: [item for item in x if item not in stop_words])
df_romance["stopwords_removed"] = df_romance["tokenized_description"].apply(lambda x: [item for item in x if item not in stop_words])
```

### 4. Get Sentiment Data

We will be using SentimentIntensityAnalyzer to obtain sentiment information for further data processing and data visualization.

```python
# Method to get sentiment information as dataframe

analyser = SentimentIntensityAnalyzer()

def get_sentiments(input_list):
    
    output = pd.DataFrame()

    for sentence in input_list:
        ss = analyser.polarity_scores(sentence)
        ss['sentence'] = sentence
        output = output.append(ss, ignore_index=True)

    return output
```

```python
# Create list of sentences from original descriptions

horror_list = list(df_horror['description'].values)
romance_list = list(df_romance['description'].values)
```

```python
# Use above method and above lists to get sentiment measures

horror_sentiments = get_sentiments(horror_list)
romance_sentiments = get_sentiments(romance_list)
```
